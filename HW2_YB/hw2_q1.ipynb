```python
def get_traffic_data(year, month,df):
    url_base = 'https://data.ibb.gov.tr/api/3/action/datastore_search?resource_id='    
    temp = df[df["date"] == f"{year}{month}"]
    dataset_id = temp['id'].values[0]
    
    url_comp= f"{url_base}{dataset_id}"
    with urlopen(url_comp) as url:
    	json_data = json.load(url)
    print(url_comp)
    df = pd.DataFrame(json_data['result']['records'])

    return df
```

```python
 df = get_traffic_data("2023", "09",df)
```

```python
df
```

```python
df_grouped = df.groupby(['GEOHASH']).agg({
        'LONGITUDE': 'first',
        'LATITUDE': 'first',
        'AVERAGE_SPEED': 'mean',
        'NUMBER_OF_VEHICLES': 'mean'
    }).reset_index()
df_grouped
```

```python
df_grouped["NUMBER_OF_VEHICLES"] = pd.to_numeric(df_grouped["NUMBER_OF_VEHICLES"])
df_grouped["AVERAGE_SPEED"] = pd.to_numeric(df_grouped["AVERAGE_SPEED"])
df_grouped["LONGITUDE"] = pd.to_numeric(df_grouped["LONGITUDE"])
df_grouped["LATITUDE"] = pd.to_numeric(df_grouped["LATITUDE"])

highest_avg_num_vehicles = df_grouped.nlargest(25, 'NUMBER_OF_VEHICLES')
highest_avg_num_vehicles
highest_avg_speed = df_grouped.nlargest(25, 'AVERAGE_SPEED')

lowest_avg_speed = df_grouped.nsmallest(25, 'AVERAGE_SPEED')
```
