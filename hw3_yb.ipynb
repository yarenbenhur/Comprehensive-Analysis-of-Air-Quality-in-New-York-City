Created a function to validate the url since we don't know how many roll call are there in a year.

```python
def is_valid_url(url):
    response = requests.head(url)
    return response.status_code == 200
   
```

```python
def generate_url(i,year):
    url_base = 'https://clerk.house.gov/evs/'
    url_year= url_base + f'{year}'

    if i<10:
            url = url_year +'/roll00'+ f'{i}' +'.xml'
    elif i<100:
            url = url_year +'/roll0'+ f'{i}' +'.xml'
    else:
            url = url_year +'/roll'+ f'{i}' +'.xml' 
    return url;
```

Url generator function with year and ending index parameter. Ending index calculated with binary search metot below.

```python
def urlGenerator(year,end_index):
    i=1;
    urlList= [];
    url_base = 'https://clerk.house.gov/evs/'
    url_year= url_base + f'{year}'
    while i!= end_index+1:
        if i<10:
            url = url_year +'/roll00'+ f'{i}' +'.xml'
        elif i<100:
            url = url_year +'/roll0'+ f'{i}' +'.xml'
        else:
            url = url_year +'/roll'+ f'{i}' +'.xml'
        urlList.append(url)
        i+=1;
    return urlList;
            
```

```python

def binary_search_valid_urls(start, end,year):
    valid_urls = []
    last_valid_index = 0;

    while start <= end:
        mid = (start + end) // 2
        current_url = generate_url(mid,year)

        if is_valid_url(current_url):
            last_valid_index = mid
            start = mid + 1
        else:
            end = mid - 1
    return last_valid_index

```

This function below gets the parsed raw for each years data.

```python
def get_raw(url_list):
    raw_list_year = [];
    for url_ in url_list:
        with urlopen(url_) as url:
            raw = xmltodict.parse(url.read())
            raw_list_year.append(raw)
    return raw_list_year

```
